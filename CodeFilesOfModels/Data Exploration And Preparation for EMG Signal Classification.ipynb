# Importing Required Libraries

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

# Reading Data

```python
train_df = pd.read_excel('trainset_60.xls', header = None)
test_df = pd.read_excel('testset_20.xls', header = None)
validate_df = pd.read_excel('validate_20.xls', header = None)
```

# Exploring Data

```python
# having a look at the first 10 rows
train_df
```

```python
# train_data analysis -> shape, columns, description and info
print("\n\nshape of the train_data (in (rows, columns) form) : ", train_df.shape)
print()
print("\n\ncolumns of the train_data : \n", train_df.columns)
print('\n\nDescription of the train_data : \n\nIt includes count, mean,standard deviation, minimum, maximum etc of each columns\n')
print(train_df.info())
print('\n\nso according to the previous result we do not have any missing or NAN data.....so we don\'t have to worry about data cleaning')
train_df.describe()
```

# Preparing Data For Classification

```python
# Converted the data frame into numpy array
train_data = np.array(train_df)
test_data = np.array(test_df)
validate_data = np.array(validate_df)

x_train = train_data[:, : train_data.shape[1] - 1]
y_train = train_data[:, train_data.shape[1] - 1]

x_validate = validate_data[:, : validate_data.shape[1] - 1]
y_validate = validate_data[:, validate_data.shape[1] - 1]

x_test = test_data[:, : test_data.shape[1] - 1]
y_test = test_data[:, test_data.shape[1] - 1]
```

```python
# Feature Scaling
scaler = StandardScaler().fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
x_validate_scaled = scaler.transform(x_validate)
```

```python
x_train_scaled_df = pd.DataFrame(x_train_scaled)
y_train_df = pd.DataFrame(y_train)

x_validate_scaled_df = pd.DataFrame(x_validate_scaled)
y_validate_df = pd.DataFrame(y_validate)

x_test_scaled_df = pd.DataFrame(x_test_scaled)
y_test_df = pd.DataFrame(y_test)
```

# Saving the scaled training Data for further use

```python
x_train_scaled_df.to_csv('x_train.csv')
x_validate_scaled_df.to_csv('x_validate.csv')
x_test_scaled_df.to_csv('x_test.csv')
```

```python
y_train_df.to_csv('y_train.csv')
y_test_df.to_csv('y_test.csv')
y_validate_df.to_csv('y_validate.csv')
```
